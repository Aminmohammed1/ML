```python
# Create a Pandas DataFrame with 3 columns and 5 rows filled with random integers
df = pd.DataFrame(np.random.randint(1, 100, size=(5, 3)), columns=['A', 'B', 'C'])
print("Original DataFrame:")
print(df)

# Introduce some NaN values
df.iloc[0, 1] = np.nan
df.iloc[2, 2] = np.nan
df.iloc[4, 0] = np.nan
print("DataFrame with NaN values:")
print(df)

# Fill the NaN values with the mean of the respective columns
df.fillna(df.mean(), inplace=True)
print("DataFrame with NaN values filled:")
print(df)
df.mean(axis=1)
```

```python
df1 = pd.DataFrame({'Key': ['A', 'B', 'C', 'D'], 'Value1': np.random.randint(1, 100, size=4)})
df2 = pd.DataFrame({'Key': ['A', 'B', 'C', 'E'], 'Value2': np.random.randint(1, 100, size=4)})
print("DataFrame 1:")
print(df1)
print("DataFrame 2:")
print(df2)

# Merge the DataFrames using the common column
merged = pd.merge(df1, df2, on='Key', how='outer')
print("Merged DataFrame:")
print(merged)
```

```python
# Create a Pandas DataFrame with a datetime index ranging from '2021-01-01' to '2021-12-31'
date_rng = pd.date_range(start='2021-01-01', end='2021-12-31', freq='D')
df = pd.DataFrame(date_rng, columns=['date'])
df['data'] = np.random.randint(0, 100, size=(len(date_rng)))
df.set_index('date', inplace=True)
print("Original DataFrame:")
print(df)

# Compute the rolling mean with a window of 7 days
rolling_mean = df.rolling(window=7).mean()
print("Rolling mean DataFrame:")
print(rolling_mean)
```

```python
data = [
    ["USA", "USA", "CAN", "CAN", "AUS"],
    ["NYC", "Toronto", "vancover", "ASC", "sydney"]
]
index = pd.MultiIndex.from_arrays(data, names=('country', "city"))

df = pd.DataFrame({'population':[10, 20, 30, 20, 11]}, index=index)
print(df)
df.loc[('USA', )]
df.xs('Toronto', level='city')
```

```python
data = [
    ["USA", "USA", "CAN", "CAN", "AUS", "USA"],
    ["NYC", "Toronto", "vancover", "ASC", "sydney", "NYC"]
]
index = pd.MultiIndex.from_arrays(data, names=('category', "subCategory"))

df = pd.DataFrame({'population':[10, 20, 30, 20, 11, 15]}, index=index)
print(df)
print(df.groupby(level='category').sum())
print(df.groupby(['category', 'subCategory']).sum())
```
