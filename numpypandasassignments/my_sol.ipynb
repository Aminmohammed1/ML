```python
dtype = np.dtype([('x', 'i4'), ('y', 'i4')])
array = np.array([(1, 2), (3, 4), (5, 6)], dtype=dtype)
# print(array)
# print(array['x'][:, np.newaxis])
print(array[:, np.newaxis])
distances = np.sqrt((array['x'][:, np.newaxis] - array['x'])**2 + (array['y'][:, np.newaxis] - array['y'])**2)
# print(array['x'])
print(distances)

# Create a structured array with fields 'x' and 'y'
data_type = [('x', 'i4'), ('y', 'i4')]
data = np.array([(1, 2), (3, 4), (5, 6)], dtype=data_type)
print("Original array:")
print(data)

# Compute the Euclidean distance between each pair of points
distances = np.sqrt((data['x'][:, np.newaxis] - data['x'])**2 + (data['y'][:, np.newaxis] - data['y'])**2)
print("Euclidean distances:")
print(distances)
```
